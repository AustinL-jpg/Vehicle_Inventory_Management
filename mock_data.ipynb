```python
#This randomizes and adds Mock data. Running this will not result in the same results as posted. 


data_mock = pd.DataFrame(data_2)

stock = []
count = 0
reset = random.randint(5, 15)

for i in data_mock['make']:
    if count == reset:
        stock.append(random.randint(1, 9))
        count = 0
        reset = random.randint(5, 15)
    else:
        stock.append(random.randint(0, 3))
    count += 1
data_mock['Stock'] = stock
data_mock.head()

vin = []
makes = []
models = []


for i, j, s, d, y in zip(data_mock['make'], data_mock['model'], data_mock['Stock'], data_mock['id'], data_mock['year']):
    make = i[0]
    model = j[0]
    make1 = i
    model1 = j
    num = s
    car_id = d
    year = y
    for row in range(num):
        number = np.random.randint(1111, 9999)
        cond = np.random.choice(['Used', 'New'])
        vin_num = f'{make}{model}{number}'
        
        if cond == 'Used':
            mileage = np.random.randint(501, 150000)
        elif cond == 'New':
            mileage = np.random.randint(0, 500)

        
        vin.append([vin_num, make1, model1, cond, mileage, num, car_id, year])

        
    

vin = pd.DataFrame(vin, columns=['VIN', 'make', 'model', 'Condition', 'Mileage', 'Stock', 'id', 'year'])
vin.head()


```
